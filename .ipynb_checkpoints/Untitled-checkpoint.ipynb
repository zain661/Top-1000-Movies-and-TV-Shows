```python
import pandas as pd
```

```python
df = pd.read_csv('imdb_top_1000.csv')
```

```python
df.drop(["Poster_Link","Runtime","Released_Year","Certificate","Overview","IMDB_Rating","Meta_score","Star1","Star2","Star3","Star4","No_of_Votes","Gross"],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Genre'].value_counts()

```

```python
df['Genre'].value_counts().head(1)
```

```python
df_2 = df['Genre'].value_counts()
```

```python
df_2
```

```python
df_2 = df['Genre'].value_counts().reset_index()
```

```python
df_2
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(50,50))
sns.barplot(x='index',y='Genre',data=df_2)
plt.xticks(rotation=90)
plt.show()
```
